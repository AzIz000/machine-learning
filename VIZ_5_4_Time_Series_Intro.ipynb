## **Plotly**

### **Initial setup**

```
#Upgrade libraries
#!pip install --upgrade plotly
import plotly
print(plotly.__version__)

#Import libraries
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid') 
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
```

### **Line**

```
flights_data = sns.load_dataset('flights')
sns.lineplot(x=flights_data.index, y='passengers', data=flights_data);
```

```
flights_data = sns.load_dataset('flights')
sns.lineplot(x='year', y='passengers', data=flights_data);
```

```
fig = px.line(flights_data, x=flights_data.index, y='passengers')
fig.show()
```

```
fig = px.line(flights_data, x=flights_data.index, y='passengers')
fig.update_xaxes(rangeslider_visible=True)
fig.show()
```

```
ma_day = [5,10,20]

for ma in ma_day:
    column_name = "MA_%s" %(str(ma))    
    flights_data[column_name] = flights_data['passengers'].rolling(window=ma,center=False).mean()

plt.figure(figsize=[12,6])

sns.lineplot(x=flights_data.index, y='MA_5', data=flights_data, label="MA_5")
sns.lineplot(x=flights_data.index, y='MA_10', data=flights_data, label="MA_10")
sns.lineplot(x=flights_data.index, y='MA_20', data=flights_data, label="MA_20")
sns.lineplot(x=flights_data.index, y='passengers', data=flights_data, label="passengers");
```

### **Bar**

```
flights_data = sns.load_dataset('flights')
sns.barplot(x='year', y='passengers', data=flights_data);
```
